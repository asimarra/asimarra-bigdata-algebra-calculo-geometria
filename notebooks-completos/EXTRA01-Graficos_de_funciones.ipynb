# Gráficos de funciones

Durante los temas anteriores hemos hecho algunas representaciones de funciones en dos o tres dimensiones. Para ello, hemos elegido ejes cartesianos, perpendiculares entre ellos y que representan (un fragmento de) la recta real.


## 1. Ejes coordenados

La recta real es una forma visual de representar un subconjunto de los números reales. Normalmente elegimos un intervalo en el que podamos representar nuestros datos cómodamente:

- Si estamos representando datos, por ejemplo, de un experimento, veremos en qué intervalo están.
- Si estamos representando una función analítica, lo que hacemos es representar su imagen en un dominio. Como entre dos números reales cualesquiera hay infinitos, también vamos a estar representando un número finito de puntos.

Los ejes cartesianos a los que estamos acostumbrados representan $\mathbb{R}²$, dos rectas reales que se cortan perpendicularme en el origen de ambas, formando un plano. Representaremos los puntos en este plano con sus coordenadas (x,y), que serán dos números reales.

## 2. Gráficos con matplotlib

Matplotlib es una librería básica de Python para dibujar gráficas, pero que tiene todo lo que necesitamos. Vamos a empezar dibujando un conjunto de puntos.

```python
x_numbers = [1, 2, 3]
y_numbers = [2, 4, 6]
```

```python
import matplotlib.pyplot as plt
plt.plot(x_numbers, y_numbers)
plt.show()
```

Los puntos que hemos elegido están alineados y matplotlib ha dibujado una recta directamente, pero podemos pedirle que marque los puntos explícitamente.

```python
plt.plot(x_numbers, y_numbers, marker='o')
plt.show()
```

```python
plt.plot(x_numbers, y_numbers, marker='*')
plt.show()
```

```python
plt.plot(x_numbers, y_numbers, marker='x')
plt.show()
```

```python
plt.plot(x_numbers, y_numbers, 'o')
plt.show()
```

## 2.1 Elementos de la gráfica

En la [documentación de matplotlib](https://matplotlib.org/stable/index.html) podemos encontrar funciones, ejemplos y casos de uso. Hay cuatro elementos básicos:

- `Figure`: Nuestra gráfica. La figura contiene los ejes y otros objetos especiales, como título, leyenda, paleta de colores... 
- `Axes`: Objeto de una figura que contiene la región en la que se dibuja, además de dos o tres objetos `Axis`. La clase Axes se utiliza cuando aplicamos OOP a las gráficas de funciones con matplotlib. 
- `Axis`: Estos objetos marcan la escala y los límites, y generan los ticks de los ejes y sus etiquetas. 
- `Artist`: Todo lo que se ve en la figura es del tipo `Artist` (incluso objetos de tipo Figure, Axes, y Axis). Cuando se renderiza la figura, se muestran todos los objetos de tipo `Artist`, la mayoría ligados a unos ejes concretos.

### 2.1.1 Creación de una gráfica con ejes

Podemos crear los ejes a la vez que la figura, o añadirlos más adelante. 

```python
np.random.seed(19680801)  
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

## 2.2 Varias gráficas en la misma figura

La librería nos permite hacer `sublpots` o subfiguras dentro de nuestra figura principal.

```python
fig = plt.figure()  # figura vacía
fig, ax = plt.subplots()  # una figura con un conjunto de ejes
fig, axs = plt.subplots(2, 2)  # una figura con 2x2 subfiguras y sus respectivos ejes
# una figura con un mosaico de subfiguras
fig, axs = plt.subplot_mosaic([['left', 'right_top'],
                               ['left', 'right_bottom']])
```

```python
fig = plt.figure(layout='constrained', facecolor='lightskyblue')
fig.suptitle('Figure')
figL, figR = fig.subfigures(1, 2)
figL.set_facecolor('thistle')
axL = figL.subplots(2, 1, sharex=True)
axL[1].set_xlabel('x [m]')
figL.suptitle('Left subfigure')
figR.set_facecolor('paleturquoise')
axR = figR.subplots(1, 2, sharey=True)
axR[0].set_title('Axes 1')
figR.suptitle('Right subfigure')
```

Veamos cómo usar cada uno de estos subplots:

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(3.5, 2.5),
                        layout="constrained")
# for each Axes, add an artist, in this case a nice label in the middle...
for row in range(2):
    for col in range(2):
        axs[row, col].annotate(f'axs[{row}, {col}]', (0.5, 0.5),
                            transform=axs[row, col].transAxes,
                            ha='center', va='center', fontsize=18,
                            color='darkgrey')
fig.suptitle('plt.subplots()')
```
